```python
#import des librairies
from lxml import etree
import collections
from xml.etree import ElementTree as et
import os
import numpy as np
import requests
import json
from collatex import *
```

```python
#on parse les documents
A1 = etree.parse('textes/lancelotaextraitC.xml')
B1 = etree.parse('textes/lancelotbextraitC.xml')
```

```python
#on charge le ns car ce sont des docs tei
tei = {'tei':'http://www.tei-c.org/ns/1.0'}
```

```python
#on récupère les lemmes grâce à xpath et en déclarant le ns : exemple :
A1.xpath("/tei:TEI/tei:text/tei:body/tei:div/tei:ab/tei:w/@lemma", namespaces=tei)
```

```python
#récupération des lemmes
vA = []
for w in A1.xpath("/tei:TEI/tei:text/tei:body/tei:div/tei:ab/tei:w", namespaces=tei):
    valL = w.get("lemma")
    print(valL)
    vA.append(valL)  
```

```python
#idem pour B
vB = []
for w in B1.xpath("/tei:TEI/tei:text/tei:body/tei:div/tei:ab/tei:w", namespaces=tei):
    valL = w.get("lemma")
    print(valL)
    vB.append(valL)  
```

```python
#on crée sous forme de texte
txt = ' '.join(vA)
#on utilise une fonction qui permet de compter le nombre de mots, après voir splité sur l'espace
counts = collections.Counter(txt.split())
motsComptesA = counts.most_common()
```

```python
#idem pour B
txtB = ' '.join(vB)
countsB = collections.Counter(txtB.split())
motsComptesB = countsB.most_common()
#on obtient une liste come suit
motsComptesB
```

```python
#on crée l'élément racine du XML dans lequel on va enregisrer les données
documentL = et.Element('donneesL')
```

```python
#pour chaque élément de la liste
for x in motsComptesA:
    print(x)
    #on crée un élément mo
    mot = et.SubElement(documentL,'mot')
    #on récupère l'index de ce sur quoi on itère
    index = motsComptesA.index(x)
    #on crée des sous-éléments lemme et occA
    lemme = et.SubElement(mot, 'lemme')
    occA = et.SubElement(mot, 'occA')
    #on récupère les valeurs de la liste : occ et lemme
    nbeOccA = motsComptesA[index][1]
    motRet = motsComptesA[index][0]
    #on met ces valeurs dans nos éléments
    lemme.text = str(motRet)
    occA.text = str(nbeOccA)
    print(occA.text)
    #on crée un sous-élément occB
    occB = et.SubElement(mot, 'occB')
    #si le mot se trouve dans la liste des B
    if motRet in vB:
        #on vérifie
        print(motRet)
        #on transforme grâce à numpy en array pour réussir à récupérer l'index de là où est le mot
        lst = np.array(motsComptesB)
        result = np.where(lst == motRet )
        #on récupère l'array et on le transforme en entier : correspond à l'index de la liste
        resultInt = result[0]
        indexB= resultInt.item()
        #on a donc l'index, et on peut récupérer l'occ
        nbeOccB = motsComptesB[indexB][1]
        #on met l'occ dans notre élément
        occB.text = str(nbeOccB)
        #si le mot n'est pas dans la liste, la valeur = 0
    else:
        occB.text= str('0')
        
```

```python
#création de la racine de l'élément xml
tree = et.ElementTree(documentL)
```

```python
##on va passer à la collation

#défintion de la fonction qui permet la transformation vers du json, puisque collatex doit prendre du json
#on utilise une feuille de transformation
def XMLtoJson(id,xmlInput):
    #défition de witness comme un dictionnaire
    witness = {}
    #on donne la première valeur de l'id
    witness['id'] = id
    #on transforme le document XML
    monXSL = etree.XML('''
<xsl:stylesheet xmlns:xsl="http://www.w3.org/1999/XSL/Transform"
    xmlns:xs="http://www.w3.org/2001/XMLSchema"
    xmlns:tei="http://www.tei-c.org/ns/1.0"
    exclude-result-prefixes="xs"
    version="1.0">
    
    <xsl:output method="text" xml:space="default"/>
    
    <xsl:template match="/">
        <xsl:apply-templates/>
    </xsl:template>
   
    
    <xsl:template match="tei:TEI">
        <xsl:apply-templates />
    </xsl:template>
     
    <xsl:template match="tei:teiHeader"/>
    
    <xsl:template match="tei:text|tei:body|tei:div|tei:ab">
        <xsl:apply-templates/>
    </xsl:template>
       
    <xsl:template match="tei:w">
        <xsl:text>{"text": "</xsl:text>
        <xsl:apply-templates/>
        <xsl:text>", "i": "</xsl:text>
        <xsl:value-of select="@xml:id"/>
        <xsl:text>", "t": "</xsl:text>
        <xsl:value-of select="@lemma"/>
        <xsl:text>", "pos": "</xsl:text>
        <xsl:value-of select="@pos"/>
        <xsl:text>"}</xsl:text>
        <xsl:if test="following::tei:w">
            <xsl:text>, </xsl:text>
        </xsl:if>
    </xsl:template>
    
    
</xsl:stylesheet>
    ''')
    #on applique l'XSL
    monXSL = etree.XSLT(monXSL)
    #on charge le résultat de l'XSL dans la valeur de l'entrée du dicitonnaire "tokens"
    witness['tokens'] = json.loads( '[' +"".join(str(monXSL(xmlInput)).split()) +']')
    return witness
```

```python
#le dictionnaire niveau général
json_input = {}
#la première entrée qui contient les witnesses
json_input['witnesses'] = []
```

```python
#application de la fonction de transformation sur A puis sur B et on met dans le dictionnaires des témoins
json_input['witnesses'].append(XMLtoJson('A',A1))
json_input['witnesses'].append(XMLtoJson('B',B1))
```

```python
#définition de la fonction notre export (comme tout à l'heure)

def notre_export_xml(collation):
    readings = []
    for column in table.columns:
        #création élément app
        app = et.Element('app')
        for key, value in sorted(column.tokens_per_witness.items()):
            child = et.Element('rdg')
            child.attrib['wit'] = "#" + key
            #on créer les attributs pour garder nos valeurs
            for item in value:
                #on ajoute la clé pour chaque id soit unique dans le XML
                child.attrib['xml:id'] = key + "_" + item.token_data["i"]
                child.attrib['lemma']= item.token_data["t"]
                child.attrib['pos'] = item.token_data["pos"]
            child.text = "".join(str(item.token_data["text"]) for item in value)
            app.append(child)
        result = et.tostring(app, encoding="unicode")
        readings.append(result)
    return "".join(readings)
```

```python
#collation sur le json créé : un peu long
json_collation = Collation()
#pour chaque témoin dans le json
for witness in json_input["witnesses"]:
#on utilise la fonction add_witness de la classe Collation
    json_collation.add_witness(witness)
    #changement de nom    
    collation = json_collation
    print(collation)
    #on collationne avec une sortie table
    table = collate(collation, output="table",segmentation=False)
    #on utilise la fonction export
    result = notre_export_xml(table)
    #on crée un élément p pour chacune des collations réalisées
    docC = '<ab>' + result + '</ab>'
    #on ajoute à la liste l'ensemble créé
   # listeDesP.append(doc)
```
